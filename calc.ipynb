```python
import pandas as pd
import numpy as np
import matplotlib
df = pd.read_csv('realtor-data.zip.csv')
```

```python
df1 = df.drop(['status','prev_sold_date'],axis=1)
```

```python
dfA = df1.groupby('state').agg('count').reset_index()
dfA["%"] = dfA['house_size']/dfA['price']
dfA
```

```python
df2 = df1[df1.state == 'New York'].dropna()
df2.drop('state',axis=1)
```

```python
df2['Price_per_sqft'] = df2['price']/df2['house_size']
df2
```

```python
location_stats = df2.groupby('city')['city'].agg('count').sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
city_10_less = location_stats[location_stats<=10]
city_10_less
```

```python
df2.city = df2.city.apply(lambda x : 'other' if x in city_10_less else x)
df2
```

```python
df2[df2.house_size/df2.bed<300].head()
```

```python
df2.shape
```

```python
df3 = df2[~(df2.house_size/df2.bed<300)]
df3.shape
```

```python
df3.Price_per_sqft.describe()
```

```python
def remove_outliers(df):
    df_return = pd.DataFrame()
    for key, sub in df.groupby('city'):
        mean = np.mean(sub.Price_per_sqft)
        std = np.std(sub.Price_per_sqft)
        cut_df = sub[(sub.Price_per_sqft>(mean - std)) & (sub.Price_per_sqft<=(mean + std))]
        df_return = pd.concat([df_return,cut_df],ignore_index=True)
    return df_return

df4 = remove_outliers(df3)
df4.shape
```

```python
def remove_bed(df):
    indicies = np.array([])
    for city, city_df in df.groupby('city'):
        beds = {}
        for bed, bed_df in city_df.groupby('bed'):
            beds[bed] = {
                'mean': np.mean(bed_df.Price_per_sqft),
                'std': np.std(bed_df.Price_per_sqft),
                'count': bed_df.shape[0]
            }
        for bed, bed_df in city_df.groupby('bed'):
            stats = beds.get(bed - 1)
            if stats and stats['count'] > 5:
                indicies = np.append(indicies, bed_df[bed_df.Price_per_sqft<(stats['mean'])].index.values)
        return df.drop(indicies,axis = 'index')

dfInsert = remove_bed(df4)
dfInsert.shape
```

```python
df5 = dfInsert[(dfInsert.bath<dfInsert.bed + 2)]
df5.shape
```

```python
df6 = df5.drop(['zip_code','Price_per_sqft','state'],axis=1)
df6
```

```python
temp = pd.get_dummies(df6.city)
```

```python
df7 = pd.concat([df6,temp.drop('other',axis='columns')],axis='columns')
df7
```

```python
df8 = df7.drop(['city'],axis=1)
df8
```

```python
X = df8.drop('price',axis =1)
X.head(2)
```

```python
y= df8.price
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train, y_train)
lr_clf.score(X_test, y_test)
```
